```python
import datetime
df_google=pd.read_csv('google.csv')
df_google_np=df_google.to_numpy()

goog_cp=df_google_np[:,4]
holidays=[datetime.datetime(2020,5,3),datetime.datetime(2021,9,3)]
date_arr=pd.bdate_range(start='9/3/2020',end='8/23/2021',
                        freq='C',holidays=holidays)
date_arr_np=date_arr.to_numpy()
fig_7=plt.figure(figsize=(8,5))
axes_7=fig_7.add_axes([0.1,0.1,0.9,0.9])
plt.plot(date_arr_np, goog_cp)
```

```python
#round to two decimal place

df_google['Open']=pd.Series([round(val,2) for val in df_google['Open']]
                           , index=df_google.index)
df_google['High']=pd.Series([round(val,2) for val in df_google['High']]
                           , index=df_google.index)
df_google['Low']=pd.Series([round(val,2) for val in df_google['Low']]
                           , index=df_google.index)
df_google['Close']=pd.Series([round(val,2) for val in df_google['Close']]
                           , index=df_google.index)
df_google['Adj Close']=pd.Series([round(val,2) for val in df_google['Adj Close']]
                           , index=df_google.index)
df_google['Volume']=pd.Series([round(val,2) for val in df_google['Volume']]
                           , index=df_google.index)
```
